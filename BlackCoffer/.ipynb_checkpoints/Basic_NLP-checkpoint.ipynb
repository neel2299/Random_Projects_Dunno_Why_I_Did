```python
from nltk.tokenize import sent_tokenize
```

```python
def accumulate_length(a):
    #print(a[3])
    acc = 0
    acc += np.mean([len(word_tokenize(sentence)) for sentence in sent_tokenize(a[3])])
    a[9] = acc
    return a
```

```python
df["words_per_sentence"] = "p"
df = df.apply(accumulate_length , axis = 1)
```
