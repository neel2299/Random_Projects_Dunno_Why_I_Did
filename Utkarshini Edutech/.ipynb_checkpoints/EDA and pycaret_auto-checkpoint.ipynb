```python
import pandas as pd
df = pd.read_csv("Churn_Modelling.csv")
```

```python
import numpy as np
```

```python
#df.info()  #no null vals
df.drop(["CustomerId","Surname","RowNumber"], axis = 1, inplace = True)
df.head()
```

```python
num_var = set([f for f in df.columns if df[f].dtype!='O'])
dis_var = set([f for f in num_var if df[f].nunique() < 10])
print(num_var, dis_var, sep = "\n")
```

```python
num_var = num_var.difference(dis_var)
num_var
```
